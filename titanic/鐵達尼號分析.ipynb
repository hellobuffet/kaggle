```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
%matplotlib inline
pd.set_option('display.max_rows',15)
```

```python
df = pd.read_csv("C:/Users/sc/titanic/train.csv")
df
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df = df.fillna(df.median())
df
```

```python
df.isnull().sum()
```

```python
df = df.drop(["Cabin"],axis = 1)
df.isnull().sum()
```

```python
df['Embarked'].value_counts()
```

```python
print("Embarked 最常出現:",df['Embarked'].value_counts().idxmax())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].value_counts().idxmax())
print("補完後的nan:")
print(df.isnull().sum())
```

```python
dummy = pd.get_dummies(df['Embarked'])
df = pd.concat([df,dummy],axis=1)
df = df.drop(["Embarked"],axis=1)
dummy = pd.get_dummies(df['Sex'])
df = pd.concat([df,dummy],axis=1)
df = df.drop(['Sex'],axis=1)
df
```

```python
s = df['Name'].str.split(",",expand=True)[1]
s = s.str.split(" ", expand = True)[1]
pd.crosstab(s,df['Survived']).T
```

```python
def name_filter(data):
    if data == 'Mr.':
        return 'Mr'
    elif data =='Mrs.':
        return 'Mrs'
    elif data == 'Miss.':
        return 'Miss'
    else:
        return "Unknown"

df['Name'] = s.apply(name_filter)

dummay = pd.get_dummies(df['Name'])
df = pd.concat([df,dummay],axis=1)
df = df.drop(['Name'],axis=1)
df = df.drop(["Unknown"],axis=1)
df
```

```python
df = df.drop(["Ticket"],axis = 1)
df = df.drop(["PassengerId"],axis = 1)
```

```python
from sklearn.model_selection import train_test_split
data_train,data_test,target_train,target_test = train_test_split(df.drop(["Survived"],axis=1),
                                                                 df['Survived'],
                                                                 test_size = 0.1)
```

```python
plt.figure(figsize=(14,10))
plt.title('Pearson Correlation of Features',y=1.05,size=15)
sns.heatmap(df.astype(float).corr(), cmap="BrBG",
            linewidths=0.1,
            square=True,
            linecolor = 'white',
            annot = True)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(max_depth=15,min_samples_leaf=20)
clf = clf.fit(data_train,target_train)
clf
```

```python
from sklearn.tree import export_graphviz
import graphviz
dot_data = export_graphviz(clf,out_file=None,
                          feature_names=df.drop(["Survived"],axis=1).columns,
                          class_names=["No","Yes"],
                          filled=True,rounded=True,
                          special_characters=True)
graph = graphviz.Source(dot_data)
graph
```

```python
from sklearn.metrics import accuracy_score
predict  = clf.predict(data_test)
print(predict)
print(list(target_test))
print(accuracy_score(target_test,predict)*100,"%")
```
